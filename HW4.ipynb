Question 5

```python
import numpy as npy
import matplotlib.pyplot as mtplt
marks={'Andy':88,'Amy':66,'James':90,'Jules':55,'Arthur':77}
x=npy.array(['Andy','Amy','James','Jules','Arthur'])
y=npy.array([88,66,90,55,77])
mtplt.bar(x,y)
mtplt.title('Grades')
mtplt.show()
```

```python
labels=['Andy','Amy','James','Jules','Arthur']
y=npy.array([88,66,90,55,77])
mtplt.pie(y,labels=labels)
mtplt.title('Grades')
mtplt.legend(bbox_to_anchor=(1.05,1.0),loc='upper left')
mtplt.show()
```

```python
import matplotlib.pyplot as mtplt
import numpy as npy
a=npy.array([1,2,3,4,5])
b=npy.array([3,7,5,9,2])
a2=npy.array([1,2,3,4,5])
b2=npy.array([2,5,8,9,10])
mtplt.subplot(2,3,1)
mtplt.plot(a,b)
mtplt.plot(a2,b2)
mtplt.title('Lines')
mtplt.show()

c=npy.array(['A','B','C','D'])
d=npy.array([3,5,7,9])
mtplt.subplot(4,5,3)
mtplt.bar(c,d)
mtplt.title('Bar Graph')
mtplt.show()

e=npy.array([10,20,50,40,30])
mtplt.subplot(3,3,4)
mtplt.pie(e)
mtplt.title('Pie')
mtplt.show()

f=npy.array([0,1,2,3,4])
g=npy.array([3,5,7,9,11])
mtplt.subplot(4,5,6)
mtplt.plot(f,g)
mtplt.grid()
mtplt.title('Grid')
mtplt.show()

h=npy.array([1,5,7,8])
i=npy.array([0,6,7,10])
mtplt.subplot(2,3,5)
mtplt.scatter(h,i,color='purple')
mtplt.title('Scatter Plot')
mtplt.show()

j=[1,2,3,5,9,5,3,8,2]
mtplt.subplot(2,3,6)
mtplt.hist(j)
mtplt.title('Histogram')
mtplt.show()
```
