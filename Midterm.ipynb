```python
txtfile=open('numbers.txt','r')
d=dict()
for nums in txtfile:
    nums=nums.strip()
    numbers=nums.split(',')
    for number in numbers:
        if number in d:
            d[number]=d[number]+1
        else:
            d[number]=1
        for key in list(d.keys()):
            print(key,':',d[key])
```

Question 2 - randomized values

```python
import statistics as stats
monthlyavg=[101.67,32.45,97.78,190.45,79.12,156.90,89.32,105.9,67.49,90.77,123.04,97.27]
print("max",(max(monthlyavg)))
print("min",(min(monthlyavg)))
print("mean",(stats.mean(monthlyavg)))
print("median",(stats.median(monthlyavg)))
print("SD",(stats.stdev(monthlyavg)))
```
